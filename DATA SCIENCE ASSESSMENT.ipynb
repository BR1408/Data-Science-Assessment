```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fp=r"C:\Users\ACER\Downloads\books.csv"
x=pd.read_csv(fp)
```

```python
r,c=x.shape
```

```python
print(r,c)
```

```python
wot=x['original_title'].isna().sum()
print(wot)
```

```python
x = x.dropna(subset=['original_title'])
nub = x['book_id'].nunique()
print(nub)
```

```python
ar=x['average_rating'].mean()
print(ar)
```

```python
biy= x[x['original_publication_year'] == 2000]
numbiy= biy.shape[0]
print(numbiy)
```

```python
book_with_max_rating = x[x['work_ratings_count'] == x['work_ratings_count'].max()]
max_ratings_book_title = book_with_max_rating['title'].values[0]

print("The book with maximum ratings is :",max_ratings_book_title)
```

```python
bins = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
x['rating_bucket'] = pd.cut(x['average_rating'], bins, labels=[str(b) for b in bins[:-1]], include_lowest=True, right=False)
bucket_counts = x['rating_bucket'].value_counts().sort_index()
plt.figure(figsize=(10,6))
bucket_counts.plot(kind='bar')
plt.xlabel('Average Rating')
plt.ylabel('Number of Books')
plt.title('Number of Books in Each Rating Bucket')
plt.show()
```
